```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.random.randint(100,size=(10000,1))
count,bins,ig=plt.hist(x,bins=50,density=True)
plt.show()
```

```python
x=np.random.normal(40,50,10000)
count,bins,ig=plt.hist(x,bins=50,density=True)
plt.show()
```

# Standard Scaling

<img src='https://wikimedia.org/api/rest_v1/media/math/render/svg/00cb9b2c9b866378626bcfa45c86a6de2f2b2e40' align="left">

```python
y=(x-np.mean(x))/np.std(x)
count,bins,ig=plt.hist(y,bins=50,density=True)
#plt.plot(bins,1/(np.sqrt(2*np.pi))*np.exp(-bins**2/2),linewidth=2, color='r')
plt.show()
```

# Mean Normalization

```python
z=(x-np.mean(x))/(np.max(x)-np.min(x))
count,bins,ig=plt.hist(z,bins=50,density=True)
plt.show()
```

*Standard Scaling and Mean normalization are ideal for models that assumes zero centric data, Ex. Principal Component Analysis(PCA)*

# Min-Max Scaling

```python
z=(x-np.min(x))/(np.max(x)-np.min(x))
count,bins,ig=plt.hist(z,bins=50,density=True)
plt.show()
```

# Unit Vector

```python
z=x/np.sqrt(np.sum(x**2))
count,bins,ig=plt.hist(z,bins=50,density=True)
plt.show()
```

*Min-max scaling and Unit Vector are ideal for features with hard boundaries*
